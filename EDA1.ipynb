```python
import pandas as pd
df = pd.read_csv("/content/Cardiotocographic.csv")
df.head(20)
```

```python
df = df.apply(pd.to_numeric)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.median(), inplace = True)
```

```python
df.isnull().sum()
```

```python
def detect_outliers(col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  return df[(df[col] < lower) | (df[col] > upper)]

outliers_counts = {}
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
for col in numeric_cols:
    outliers_counts[col] = len(detect_outliers(col))
print(pd.DataFrame(outliers_counts, index =["outliers"]).T)
```

```python
for col in numeric_cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  df[col] = df[col].apply(lambda x: lower if x < lower else upper if x > upper else x)
```

```python
outlier_counts_after = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_counts_after[col] = len(outliers)

print(pd.DataFrame(outlier_counts_after, index=["Remaining Outliers"]).T)
```

**Key Insights from Exploratory Analysis**

LB (Baseline FHR), ASTV, and MLTV show meaningful variation and may be useful predictive features.

DS and DP have zero variance  they contain only zeros and can be removed from the dataset.

Several variables (AC, FM, DL, ALTV) are highly skewed, meaning the events they represent are rare.

Some medical variables contain negative values, which are not physiologically valid and require correction.

The target variable NSP contains only one class, making classification or supervised learning impossible in its current form.

Features like ASTV, ALTV, Width show high spread and may be strong indicators of fetal stress or abnormality.

**Impact on Decision-Making & Further Analysis**

Zero-variance features should be removed to reduce noise and improve model efficiency.

Skewed variables may need transformation or resampling before being used in machine learning.

Negative values must be cleaned or clipped to medically valid ranges.

The dataset must be checked for missing NSP classes before building classification models.

Features with high variability should be prioritized for feature selection and importance analysis.

Additional preprocessing is required before reliable clinical interpretation or predictive modeling.
